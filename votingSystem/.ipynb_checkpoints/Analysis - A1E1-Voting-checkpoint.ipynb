```python
import re
from collections import defaultdict
import numpy as np  # For efficient column-wise mean calculation

# Extract recalls for coordinators
def recall_extract_numbers_by_coordinator(filename):
    data = defaultdict(list)  # Dictionary to store numbers per coordinator
    pattern = r"Split \d+: Recall\s+Coordinator\s+(\d+) => \[?([\d\.]+)\]?"
    
    with open(filename, 'r') as file:
        for line in file:
            match = re.search(pattern, line)
            if match:
                coordinator = int(match.group(1))  # Extract Coordinator number
                value = float(match.group(2))      # Extract floating-point value
                data[coordinator].append(value)

    return data
    
# Extract f1 scores for coordinators
def f1_score_extract_numbers_by_coordinator(filename):
    data = defaultdict(list)  # Dictionary to store numbers per coordinator
    pattern = r"Split \d+: F1\s+Coordinator\s+(\d+) => \[?([\d\.]+)\]?"
    
    with open(filename, 'r') as file:
        for line in file:
            match = re.search(pattern, line)
            if match:
                coordinator = int(match.group(1))  # Extract Coordinator number
                value = float(match.group(2))      # Extract floating-point value
                data[coordinator].append(value)

    return data


# Extract recalls from models
def recall_extract_numbers_by_coordinator_models(filename):
    data = defaultdict(list)  # Dictionary to store lists of values per coordinator
    pattern = r"Split \d+: Recall\s+Models\s+Coordinator\s+(\d+) => \[([\d\.,\s]+)\]"

    with open(filename, 'r') as file:
        for line in file:
            match = re.search(pattern, line)
            if match:
                coordinator = int(match.group(1))  # Extract Coordinator number
                values = list(map(float, match.group(2).split(',')))  # Convert list elements to floats
                data[coordinator].append(values)
    return data

# Extract f1_score from models
def f1_score_extract_numbers_by_coordinator_models(filename):
    data = defaultdict(list)  # Dictionary to store lists of values per coordinator
    pattern = r"Split \d+: F1\s+Models\s+Coordinator\s+(\d+) => \[([\d\.,\s]+)\]"

    with open(filename, 'r') as file:
        for line in file:
            match = re.search(pattern, line)
            if match:
                coordinator = int(match.group(1))  # Extract Coordinator number
                values = list(map(float, match.group(2).split(',')))  # Convert list elements to floats
                data[coordinator].append(values)
    return data

def calculate_means(data):
    means = {coord: sum(values) / len(values) for coord, values in data.items() if values != 0}
    return means
    
def calculate_column_means(data):
    means = {}

    for coordinator, lists in data.items():
        # Transpose list of lists to group values by column
        num_columns = len(lists[0])  # Assuming all rows have the same number of columns
        column_sums = [0] * num_columns
        column_counts = [0] * num_columns

        for row in lists:
            for i, value in enumerate(row):
                if value != 0:  # Ignore zeros
                    column_sums[i] += value
                    column_counts[i] += 1

        # Compute mean for each column (avoid division by zero)
        column_means = [
            (column_sums[i] / column_counts[i]) if column_counts[i] > 0 else 0
            for i in range(num_columns)
        ]

        means[coordinator] = column_means  # Store results

    return means
    

def calculate_overall_column_mean(mean_values):
    num_columns = len(next(iter(mean_values.values())))  # Get number of columns from first coordinator
    overall_means = []

    for col in range(num_columns):
        column_values = [means[col] for means in mean_values.values() if means[col] != 0]  # Ignore 0s
        mean_value = sum(column_values) / len(column_values) if column_values else 0  # Avoid division by zero
        overall_means.append(mean_value)

    return overall_means
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from collections import defaultdict

def plot(title, data_coordinators, data_models):
    label_mapping = {1: "Mean", 3: "W. Mean", 5: "Mode"}
    
    for coordinator in data_coordinators.keys():
        plt.figure(figsize=(7, 4))  # Create a new figure for each coordinator
        
        # Convert coordinator's data to a DataFrame
        df_coordinator = pd.DataFrame({
            "Step": range(len(data_coordinators[coordinator])),
            "Recall": data_coordinators[coordinator],
            "Type": label_mapping[coordinator]
        })
    
        # Convert model data to a DataFrame and Melt df_models to long format for Seaborn plotting
        df_models = pd.DataFrame(data_models[coordinator], columns=[f"Model {i+1}" for i in range(5)])
        df_models["Step"] = df_models.index  # Add a time step index
        df_models_long = df_models.melt(id_vars=["Step"], var_name="Model", value_name="Recall")
    
        # Plot the coordinator data
        sns.lineplot(data=df_coordinator, x="Step", y="Recall", label=f"Coordinator {coordinator}", color="black", linewidth=2, marker="o")
    
        # Plot model predictions
        sns.lineplot(data=df_models_long, x="Step", y="Recall", hue="Model", palette="coolwarm", linestyle="dashed")
    
        # Add labels and title
        plt.xlabel("Time Step")
        plt.ylabel("Recall")
        plt.title(f"{label_mapping[coordinator]} - {title}")
        plt.legend(title="Legend")
        
        # Show the plot
        plt.show()
```

```python
# F1 score and Recall results for A1, E1 without attack - All 

print("Results for coordinators Analysis 1, Experiment 1 - Without attack")
filename_A1E1 = "experiments/A1E1_noatt_f1.out"
# Recall results for A1, E1 without attack - Voting
print("Results for models Analysis 1, Experiment 1 - Without attack")
recall_data_coordinators = recall_extract_numbers_by_coordinator(filename_A1E1)
f1_score_data_coordinators = f1_score_extract_numbers_by_coordinator(filename_A1E1)
recall_coordinators_mean_values = recall_calculate_means(data_coordinators)
f1_score_coordinators_mean_values = f1_score_calculate_means(data_coordinators)

recall_data_models = recall_extract_numbers_by_coordinator_models_voting(filename_A1E1)
f1_score_data_models = f1_score_extract_numbers_by_coordinator_models_voting(filename_A1E1)
recall_models_mean_values = calculate_column_means(recall_data_models)
f1_score_models_mean_values = calculate_column_means(f1_score_data_models)

# Print results f1_score
print("F1_score Coordinators Voting")
for coordinator, means in sorted(f1_score_coordinators_mean_values.items()):
    print(f"Coordinator {coordinator}: Column-wise Means = {means}")
print("F1_score Models Voting")
for coordinator, means in sorted(f1_score_models_mean_values.items()):
    print(f"Coordinator {coordinator}: Column-wise Means = {means}")
print(f"F1_score: Overall mean values for models voting: {calculate_overall_column_mean(f1_score_models_mean_values)}")
print("")

# Print results recall
print("Recall Coordinators Voting")
for coordinator, means in sorted(recall_coordinators_mean_values.items()):
    print(f"Coordinator {coordinator}: Column-wise Means = {means}")
print("Recall Models Voting")
for coordinator, means in sorted(recall_models_mean_values.items()):
    print(f"Coordinator {coordinator}: Column-wise Means = {means}")

print(f"Recall: Overall mean values for models voting: {calculate_overall_column_mean(recall_models_mean_values)}")
print("")

plot("F1 score: Model Predictions vs Collaborative predictions", f1_score_data_coordinators, f1_score_data_models)
plot("Recall: Model Predictions vs Collaborative predictions", recall_data_coordinators, recall_data_models)
```
