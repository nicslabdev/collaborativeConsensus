```python
def trim_array_to_equal_parts(arr, num_parts):
    # Calculate the maximum number of elements that can be equally divided
    total_elements = len(arr)
    divisible_length = total_elements - (total_elements % num_parts)
    
    # Trim the array to make it divisible by num_parts
    return arr[:divisible_length]

# Train Data
num_splits = 50
x_splits_train_s1 = np.split(trim_array_to_equal_parts(x_train_s1, num_splits), num_splits) 
y_splits_train_s1 = np.split(trim_array_to_equal_parts(y_train_s1, num_splits), num_splits)
print("Shape of train data splits.")
print(x_splits_train_s1[num_splits - 1].shape)
print(y_splits_train_s1[num_splits - 1].shape)


for f_idx, feature in enumerate(features):
    if f_idx >= 0:
        # Convertir en DataFrame
        data_list = []
        for i in range(50):  # 50 segmentos
            data_list.append(pd.DataFrame({'Valor': x_splits_train_s1[i][:, f_idx], 'Tipo': 'Normal', 'Split': i}))
    
        df = pd.concat(data_list, ignore_index=True)
    
        # Graficar boxplot por segmento
        plt.figure(figsize=(12, 6))
        sns.boxplot(x='Split', y='Valor', hue='Tipo', data=df, showfliers=False)
        plt.xticks(rotation=90)
        plt.title(f"Distribution of {feature} feature")
        plt.savefig(f"{plot_path_s1}/train_data_box_{feature}.png", dpi=300, bbox_inches="tight") 
        #plt.show()
```

```python
# Boulder
x_boulder = boulderData.loc[:, features].values
y_boulder = boulderData.loc[:, ['anomaly']].values.ravel()
size = len(x_boulder)
size_init = int(size * 0.8)
x_boulder_init = x_boulder[:size_init]
y_boulder_init = y_boulder[:size_init]
x_boulder_inc = x_boulder[size_init:]
y_boulder_inc = y_boulder[size_init:]

# Dundee
x_dundee = dundeeData.loc[:, features].values
y_dundee = dundeeData.loc[:, ['anomaly']].values.ravel()
size = len(x_dundee)
size_init = int(size * 0.8)
x_dundee_init = x_dundee[:size_init]
y_dundee_init = y_dundee[:size_init]
x_dundee_inc = x_dundee[size_init:]
y_dundee_inc = y_dundee[size_init:]

# Netherlands
x_netherlands = netherlandsData.loc[:, features].values
y_netherlands = netherlandsData.loc[:, ['anomaly']].values.ravel()
size = len(x_netherlands)
size_init = int(size * 0.8)
x_netherlands_init = x_netherlands[:size_init]
y_netherlands_init = y_netherlands[:size_init]
x_netherlands_inc = x_netherlands[size_init:]
y_netherlands_inc = y_netherlands[size_init:]

# Paloalto
x_paloalto = paloaltoData.loc[:, features].values
y_paloalto = paloaltoData.loc[:, ['anomaly']].values.ravel()
size = len(x_paloalto)
size_init = int(size * 0.8)
x_paloalto_init = x_paloalto[:size_init]
y_paloalto_init = y_paloalto[:size_init]
x_paloalto_inc = x_paloalto[size_init:]
y_paloalto_inc = y_paloalto[size_init:]

# Perth
x_perth = perthData.loc[:, features].values
y_perth = perthData.loc[:, ['anomaly']].values.ravel()
size = len(x_perth)
size_init = int(size * 0.8)
x_perth_init = x_perth[:size_init]
y_perth_init = y_perth[:size_init]
x_perth_inc = x_perth[size_init:]
y_perth_inc = y_perth[size_init:]
```

```python
boulderData = boulderData.sort_values(by='startTimestamp', ascending=True)
dundeeData = dundeeData.sort_values(by='startTimestamp', ascending=True)
netherlandsData = netherlandsData.sort_values(by='startTimestamp', ascending=True)
paloaltoData = paloaltoData.sort_values(by='startTimestamp', ascending=True)
perthData = perthData.sort_values(by='startTimestamp', ascending=True)
encode_variable(boulderData, boulderData.columns.get_loc("connectorType"))
encode_variable(dundeeData, dundeeData.columns.get_loc("connectorType"))
encode_variable(netherlandsData, netherlandsData.columns.get_loc("connectorType"))
encode_variable(paloaltoData, paloaltoData.columns.get_loc("connectorType"))
encode_variable(perthData, perthData.columns.get_loc("connectorType"))

x_inc_s2 = np.concatenate((x_boulder_inc, x_netherlands_inc), 0)
x_inc_s2 = np.concatenate((x_inc_s2, x_dundee_inc), 0)
x_inc_s2 = np.concatenate((x_inc_s2, x_paloalto_inc), 0)
x_inc_s2 = np.concatenate((x_inc_s2, x_perth_inc), 0)

y_inc_s2 = np.concatenate((y_boulder_inc, y_netherlands_inc), 0)
y_inc_s2 = np.concatenate((y_inc_s2, y_dundee_inc), 0)
y_inc_s2 = np.concatenate((y_inc_s2, y_paloalto_inc), 0)
y_inc_s2 = np.concatenate((y_inc_s2, y_perth_inc), 0)

# Test Data S2
num_splits = 50
x_splits_test_s2 = np.split(trim_array_to_equal_parts(x_inc_s2, num_splits), num_splits) 
y_splits_test_s2 = np.split(trim_array_to_equal_parts(y_inc_s2, num_splits), num_splits)
print("Shape of test data splits S2.")
print(x_splits_test_s2[num_splits - 1].shape)
print(y_splits_test_s2[num_splits - 1].shape)


x_adv_s2 = adversarial_data_s2.loc[:, features].values
y_adv_s2 = adversarial_data_s2.loc[:, ['anomaly']].values.ravel()
x_test_adversarial = np.array(x_adv_s2, dtype=np.float32) # Datos anomalos adversarios
y_test_adversarial = np.array(y_adv_s2, dtype=np.float32) # hay que mezclar con datos normales

anomalous_indices_s2 = np.where(y_dundee_inc == 1)[0]
X_modified_s2, Y_modified_s2  = x_dundee_inc.copy(), y_dundee_inc.copy()
X_modified_s2[anomalous_indices_s2] = x_test_adversarial
Y_modified_s2[anomalous_indices_s2] = y_test_adversarial

x_inc_adv_s2 = np.concatenate((x_boulder_inc, x_netherlands_inc), 0)
x_inc_adv_s2 = np.concatenate((x_inc_adv_s2, X_modified_s2), 0)
x_inc_adv_s2 = np.concatenate((x_inc_adv_s2, x_paloalto_inc), 0)
x_inc_adv_s2 = np.concatenate((x_inc_adv_s2, x_perth_inc), 0)

y_inc_adv_s2 = np.concatenate((y_boulder_inc, y_netherlands_inc), 0)
y_inc_adv_s2 = np.concatenate((y_inc_adv_s2, Y_modified_s2), 0)
y_inc_adv_s2 = np.concatenate((y_inc_adv_s2, y_paloalto_inc), 0)
y_inc_adv_s2 = np.concatenate((y_inc_adv_s2, y_perth_inc), 0)

# Test adversarial data
x_splits_adv_s2 = np.split(trim_array_to_equal_parts(x_inc_adv_s2, num_splits), num_splits) 
y_splits_adv_s2 = np.split(trim_array_to_equal_parts(y_inc_adv_s2, num_splits), num_splits)
print("Shape of adv data splits.")
print(x_splits_adv_s2[num_splits - 1].shape)
print(y_splits_adv_s2[num_splits - 1].shape)
```
