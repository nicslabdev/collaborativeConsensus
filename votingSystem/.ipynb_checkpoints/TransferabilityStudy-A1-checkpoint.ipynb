```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, TensorDataset
import torchattacks
from torchattacks import FGSM, PGD, MIFGSM 
import pandas as pd
from functions import encode_variable
from sklearn.model_selection import train_test_split
import numpy as np
```

```python
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(device)
```

```python
class SurrogateModel(nn.Module):
    def __init__(self, input_dim):
        super(SurrogateModel, self).__init__()
        self.fc1 = nn.Linear(input_dim, 128) 
        self.fc2 = nn.Linear(128, 64)
        self.fc3 = nn.Linear(64, 2)  # Binary classification (Normal vs. Anomalous)

    def forward(self, x):
        x = torch.relu(self.fc1(x))
        x = torch.relu(self.fc2(x))
        return self.fc3(x)
```

```python
# SAME DATASET
data = pd.read_excel("../datasets/final/Dundee_Clean_With_Anomalies_All.xlsx")
data = data.sort_values(by='startTimestamp', ascending=True)
data = encode_variable(data, data.columns.get_loc("connectorType"))
```

```python
features = ["connectorType", "durationSession", "durationCharge", "energy", 'cost', 'tariff', "meanPower", 'maxPower']

x_data = data.loc[:, features].values
y_data = data.loc[:, ['anomaly']].values.ravel()

size = len(x_data)
size_init = int(size * 0.8)
x_train = x_data[:size_init]
y_train = y_data[:size_init]
x_test = x_data[size_init:]
y_test = y_data[size_init:]


#x_train, x_test, y_train, y_test = train_test_split(x_data, y_data, test_size=0.2, shuffle=True)
x_train = np.array(x_train, dtype=np.float32) 
x_test = np.array(x_test, dtype=np.float32)

from imblearn.under_sampling import RandomUnderSampler
# Resample your dataset
rus = RandomUnderSampler(sampling_strategy='auto')
X_train_resampled, y_train_resampled = rus.fit_resample(x_train, y_train)
```

```python
X_train = torch.tensor(X_train_resampled, dtype=torch.float32)
y_train = torch.tensor(y_train_resampled, dtype=torch.long)
```

```python
# TRAINING SURROGATE MODEL
input_dim = X_train.shape[1]
model_surrogate = SurrogateModel(input_dim).to(device)
optimizer = optim.Adam(model_surrogate.parameters(), lr=0.001)
criterion = nn.CrossEntropyLoss()
train_loader = DataLoader(TensorDataset(X_train, y_train), batch_size=32, shuffle=True)
for epoch in range(10):  # Train for a few epochs
    for X_batch, y_batch in train_loader:
        X_batch, y_batch = X_batch.to(device), y_batch.to(device)
        optimizer.zero_grad()
        outputs = model_surrogate(X_batch)
        loss = criterion(outputs, y_batch)
        loss.backward()
        optimizer.step()

print("Surrogate model training complete.")
```

```python
# TESTING SURROGATE MODEL
from sklearn.metrics import accuracy_score, f1_score
# Test surrogate model
model_surrogate.eval()  # Set the model to evaluation mode

x_test_surrogate = torch.tensor(x_test, dtype=torch.float32).to(device)
y_test_surrogate = torch.tensor(y_test, dtype=torch.long).to(device)

with torch.no_grad():
    y_pred = model_surrogate(x_test_surrogate).argmax(axis=1).cpu().numpy()
# Calculate Accuracy
accuracy = accuracy_score(y_test_surrogate.cpu().numpy(), y_pred)

# Calculate Recall
from sklearn.metrics import recall_score
recall = recall_score(y_test_surrogate.cpu().numpy(), y_pred)

# Calculate F1 Score
f1 = f1_score(y_test_surrogate.cpu().numpy(), y_pred, average='weighted')  # 'weighted' handles imbalanced classes

original_results = {
    "Accuracy": accuracy, 
    "F1 Score": f1,
    "Recall": recall
}
print(f"Original results: {original_results}")
```

```python
# FUNCTION TO EXTRACT ONLY ANOMALOUS DATA
def load_anomalous_test_samples(x_test, y_test, device):
    # Select only anomalous samples (where label == 1)
    anomalous_indices = np.where(y_test == 1)[0]  # Get indices of anomalous samples
    X_anomalous = x_test[anomalous_indices]
    y_anomalous = y_test[anomalous_indices]

    # Ensure correct dtype
    X_anomalous_tensor = torch.tensor(X_anomalous, dtype=torch.float32).to(device)
    #y_anomalous_tensor = torch.tensor(y_anomalous, dtype=torch.long)
    y_anomalous_tensor = y_anomalous

    #return X_anomalous_tensor, y_anomalous_tensor
    return X_anomalous, y_anomalous, anomalous_indices
# FUNCTION TO EXTRACT ONLY NORMAL DATA
def load_normal_test_samples(x_test, y_test, device):
    # Select only anomalous samples (where label == 1)
    normal_indices = np.where(y_test == 0)[0]  # Get indices of anomalous samples
    X_normal = x_test[normal_indices]
    y_normal = y_test[normal_indices]

    # Ensure correct dtype
    X_normal_tensor = torch.tensor(X_normal, dtype=torch.float32).to(device)
    #y_anomalous_tensor = torch.tensor(y_anomalous, dtype=torch.long)
    y_normal_tensor = y_normal

    #return X_anomalous_tensor, y_anomalous_tensor
    return X_normal, y_normal
```

```python
# Adding Mask, and reality to values
class RealTabularMIFGSM(torchattacks.MIFGSM):
    def __init__(self, model, eps=0.03, alpha=0.007, steps=10, decay=1.0, mask=None, feature_min = None, feature_max = None):
        super().__init__(model, eps, alpha, steps, decay)
        self.loss = nn.CrossEntropyLoss()  # ✅ Define the loss function
        self.mask = mask  # Store the feature mask
        self.feature_min = feature_min
        self.feature_max = feature_max
        
    def forward(self, images, labels):
        """Overrides MIFGSM for tabular data."""
        
        adv_images = images.clone().detach()
        adv_images.requires_grad = True 
        
        momentum = 0
        for _ in range(self.steps):
            outputs = self.model(adv_images)
            loss = self.loss(outputs, labels)

            # Debugging statements
            #print("Loss value:", loss.item())
            #print("Loss requires_grad:", loss.requires_grad)
            
            # Compute gradients of loss with respect to the input
            grad = torch.autograd.grad(loss, adv_images, retain_graph=True, create_graph=True, allow_unused=True)[0]

            if grad is None:
                print("ERROR: Gradient is None! adv_images is not in the graph!")
                break  # Stop execution if something is wrong

            # Normalize and update adversarial images
            grad = grad / torch.mean(torch.abs(grad), dim=(1,), keepdim=True)
            grad = grad + momentum * self.decay
            momentum = grad

            # Apply mask: Only modify unprotected features
            perturbation = self.alpha * grad.sign()
            if self.mask is not None:
                perturbation = perturbation * self.mask  # Zero out protected features

            adv_images = adv_images + perturbation
            adv_images = torch.clamp(adv_images, min=self.feature_min, max=self.feature_max)  # Ensure valid range

            #adv_images.requires_grad = True  # Keep gradients enabled after update
    
        return adv_images
```

```python
# Make attack more real

# Analyse maximum and minimum values. Extract categorical features.
def extract_column_stats(df):
    stats = {}
    
    for column in df.columns:
        if pd.api.types.is_numeric_dtype(df[column]):
            stats[column] = {
                'max': df[column].max(),
                'min': df[column].min(),
                'mean': df[column].mean()
            }
        else:
            value_counts = df[column].value_counts()
            stats[column] = {
                'unique_values': value_counts.index.tolist(),
                'occurrences': value_counts.tolist()
            }
    
    return stats

stats = extract_column_stats(data.loc[:, features])
print(stats)

# Features are: "connectorType", "durationSession", "durationCharge", "energy", 'cost', 'tariff', "meanPower", 'maxPower'
# Define the minimum and maximum values for each feature
feature_min = torch.tensor([0, -1,  -1, -76,  -59,  0, -375, 7], dtype=torch.float32).to(device)  # Min values
feature_max = torch.tensor([3, 2280, 1074, 72882, 1230,  76,  417310, 150], dtype=torch.float32).to(device)  # Max values

# Define a binary mask (1 = modifiable, 0 = protected)
mask = torch.tensor([0, 1, 1, 1, 1, 1, 1, 1], dtype=torch.float32).to(device)  # Feature 2 and 6 are protected
```

```python
def save_samples_csv(adversarial_samples, eps):
    adversarial_dataset_MIFGSM = pd.DataFrame(adversarial_samples.detach().cpu().numpy(), columns=data.loc[:, features].columns)
    adversarial_dataset_MIFGSM['anomaly'] = y.detach().cpu().numpy()
    advPath = f'./adv_examples/adversarialexamplesMIFGSM_eps{str(eps)}.csv'
    adversarial_dataset_MIFGSM.to_csv(advPath, index=False)
    print(f"{advPath} converted to CSV")
    
def adversarial_attack(eps, alpha, steps, x, y, mask, feature_min, feature_max):
    attackMIFGSM = RealTabularMIFGSM(model_surrogate, eps=eps, alpha=alpha, steps=steps, decay=0.8, mask=mask, feature_min=feature_min,
                                     feature_max=feature_max)
    adversarial_samples = attackMIFGSM(x, y)
    save_samples_csv(adversarial_samples, eps) # Save in CSV for transfer
```

```python
# GENERATE ADVERSARIAL EXAMPLES
steps = 10
epsilons = [0.0]
# First loop: from 0.003 to 0.03, step = 0.003
for value in np.arange(0.003, 0.031, 0.006): 
    eps = round(value, 3)
    alpha = round(eps/steps, 4)
    adversarial_attack(eps, alpha, steps, x, y, mask, feature_min, feature_max)
    epsilons.append(eps)

# Second loop: from 0.03 to 0.3, step = 0.02
for value in np.arange(0.03, 0.3, 0.06):
    eps = round(value, 3)
    alpha = round(eps/steps, 4)
    adversarial_attack(eps, alpha, steps, x, y, mask, feature_min, feature_max)
    epsilons.append(eps)
```

```python
from sklearn.metrics import accuracy_score, f1_score
from sklearn.metrics import recall_score

def read_adv_data(advPath):
    adversarial_data = pd.read_csv(advPath)
    adversarial_x_data = adversarial_data.loc[:, features].values
    adversarial_y_data = adversarial_data.loc[:, ['anomaly']].values.ravel()

    return adversarial_x_data, adversarial_y_data 
    
def test_adversarial_samples(advPath, x_test, y_test, model_surrogate, anomalous_indices):
    # TEST ADVERSARIAL SAMPLES AND PERFORM ADVERSARIAL ATTACK
    features = ["connectorType", "durationSession", "durationCharge", "energy", 'cost', 'tariff', "meanPower", 'maxPower']

    adversarial_x_data, adversarial_y_data = read_adv_data(advPath)

    X_modified, Y_modified  = x_test.copy(), y_test.copy()
    X_modified[anomalous_indices] = adversarial_x_data
    Y_modified[anomalous_indices] = adversarial_y_data
    #X_normal_test, y_normal_test = load_normal_test_samples(x_test, y_test, device) # Para probar la performance con y normal y adversaria
    #X_normal_test = X_normal_test[:X_anomalous_test.shape[0], :]
    #y_normal_test = y_normal_test[:y_anomalous_test.shape[0]]

    #x_fused = np.concatenate((adversarial_x_data, X_normal_test), axis=0)  # Stack row-wise
    #y_fused = np.concatenate((adversarial_y_data, y_normal_test), axis=0)  

    model_surrogate.eval()
    tensor_x_fused = torch.tensor(np.array(X_modified, dtype=np.float32), dtype=torch.float32).to(device)
    tensor_y_fused = torch.tensor(np.array(Y_modified, dtype=np.float32), dtype=torch.long).to(device)
    
    with torch.no_grad():
        y_pred = model_surrogate(tensor_x_fused).argmax(axis=1).cpu().numpy()

    accuracy = accuracy_score(Y_modified, y_pred)
    recall = recall_score(Y_modified, y_pred)
    f1 = f1_score(Y_modified, y_pred)

    return  accuracy, f1, recall
X_modified, Y_modified  = x_test.copy(), y_test.copy()
```

```python
try_epsilon = [0.0, 0.003, 0.021, 0.090, 0.210] # different epsilons from less to more aggresive attacks
metrics = ["Accuracy", "F1 Score", "Recall", "EIR"]
# Create MultiIndex for rows (5 models × 3 metrics each)
multi_index = pd.MultiIndex.from_product([model_names, metrics], names=["Model", "Metric"])
df_results = pd.DataFrame(index=multi_index, columns=try_epsilon)

# Refill epsilon 0 for models
# Fill in original results for epsilon=0.0
for idx, model in enumerate(models):
    model_name = model_names[idx]
    df_results.loc[(model_name, "Accuracy"), 0.0] = original_results[model_name]["Accuracy"]
    df_results.loc[(model_name, "F1 Score"), 0.0] = original_results[model_name]["F1 Score"]
    df_results.loc[(model_name, "Recall"), 0.0] = original_results[model_name]["Recall"]
    df_results.loc[(model_name, "EIR"), 0.0] = "-"
 

for idx, model in enumerate(models):
    model_name = model_names[idx]
    for eps in try_epsilon[1:]:
        adv_path = f'./adv_examples/adversarialexamplesMIFGSM_eps{str(eps)}.csv'
        adversarial_x_data, adversarial_y_data = read_adv_data(adv_path)
        x_test_adversarial = np.array(adversarial_x_data, dtype=np.float32)
        y_test_adversarial = np.array(adversarial_y_data, dtype=np.float32)

        X_normal_test = x_test[:x_test_adversarial.shape[0], :]
        y_normal_test = y_test[:y_test_adversarial.shape[0]]
        
        x_fused = np.concatenate((x_test_adversarial, X_normal_test), axis=0)  # Stack row-wise
        y_fused = np.concatenate((y_test_adversarial, y_normal_test), axis=0) 

        preds_adv = model.predict(x_fused)
        
        df_results.loc[(model_name, "Accuracy"), eps] = accuracy_score(y_fused, preds_adv)
        df_results.loc[(model_name, "F1 Score"), eps] = f1_score(y_fused, preds_adv)
        df_results.loc[(model_name, "Recall"), eps] = recall_score(y_fused, preds_adv)   
        eir = 1 - (df_results.loc[(model_name, "Recall"), eps] / original_results[model_name]["Recall"])
        df_results.loc[(model_name, "EIR"), eps] = eir * 100

    
df_results    
```

```python
def test_adversarial_samples_s2(x_test, y_test, x_dundee, y_dundee, models_a2, anomalous_indices):
    # TEST ADVERSARIAL SAMPLES AND PERFORM ADVERSARIAL ATTACK
    features = ["connectorType", "durationSession", "durationCharge", "energy", 'cost', 'tariff', "meanPower", 'maxPower']
    df_results = pd.DataFrame(index=multi_index, columns=try_epsilon)
    
    # Refill epsilon 0 for models
    # Fill in original results for epsilon=0.0
    for idx, model in enumerate(models_a2):
        model_name = model_names[idx]
        df_results.loc[(model_name, "Accuracy"), 0.0] = original_results[model_name]["Accuracy"]
        df_results.loc[(model_name, "F1 Score"), 0.0] = original_results[model_name]["F1 Score"]
        df_results.loc[(model_name, "Recall"), 0.0] = original_results[model_name]["Recall"]
        df_results.loc[(model_name, "EIR"), 0.0] = "-"

    for idx, model in enumerate(models_a2):
        model_name = model_names[idx]
        for eps in try_epsilon[1:]:
            adv_path = f'./adv_examples_a2/adversarialexamplesMIFGSM_eps{str(eps)}.csv'
            adversarial_x_data, adversarial_y_data = read_adv_data(adv_path)
            x_test_adversarial = np.array(adversarial_x_data, dtype=np.float32)
            y_test_adversarial = np.array(adversarial_y_data, dtype=np.float32)

            #anomalous_indices = = np.where(y_dundee == 1)[0]
            X_modified, Y_modified  = x_dundee.copy(), y_dundee.copy()
            X_modified[anomalous_indices] = x_test_adversarial
            Y_modified[anomalous_indices] = y_test_adversarial
            
            x_fused = np.concatenate((X_modified, x_test), axis=0)
            y_fused = np.concatenate((Y_modified, y_test), axis=0) 
    
            preds_adv = model.predict(x_fused)
            
            df_results.loc[(model_name, "Accuracy"), eps] = accuracy_score(y_fused, preds_adv)
            df_results.loc[(model_name, "F1 Score"), eps] = f1_score(y_fused, preds_adv)
            df_results.loc[(model_name, "Recall"), eps] = recall_score(y_fused, preds_adv)   
            eir = 1 - (df_results.loc[(model_name, "Recall"), eps] / original_results[model_name]["Recall"])
            df_results.loc[(model_name, "EIR"), eps] = eir * 100

    return  df_results
```
