```python
filename1 = "nohup-test.out"  
filename2 = "nohup_mainBoth_v2-test.out"
```

```python
import re

def extract_numbers_from_file(filename):
    numbers = []
    pattern = r"Split \d+: F1 Coordinator \d+ => ([\d\.]+)"
    
    with open(filename, 'r') as file:
        for line in file:
            match = re.search(pattern, line)
            if match:
                numbers.append(float(match.group(1)))  # Convert extracted number to float

    return numbers

def calculate_mean(numbers):
    return sum(numbers) / len(numbers) if numbers else 0

# Usage

numbers = extract_numbers_from_file(filename2)
mean_value = calculate_mean(numbers)

print(f"Extracted Numbers: {numbers}")
print(f"Mean: {mean_value}")
```
