```python
import re
from collections import defaultdict
import numpy as np  # For efficient column-wise mean calculation

# Extract recalls for coordinators
def extract_numbers_by_coordinator(filename):
    data = defaultdict(list)  # Dictionary to store numbers per coordinator
    pattern = r"Split \d+: (?:F1|Recall)\s+Coordinator\s+(\d+) => ([\d\.]+)"
    
    with open(filename, 'r') as file:
        for line in file:
            match = re.search(pattern, line)
            if match:
                coordinator = int(match.group(1))  # Extract Coordinator number
                value = float(match.group(2))      # Extract floating-point value
                data[coordinator].append(value)

    return data
def calculate_means(data):
    means = {coord: sum(values) / len(values) for coord, values in data.items() if values != 0}
    return means

            # Extract recalls for coordinators voting#
def extract_numbers_by_coordinator_voting(filename):
    data = defaultdict(list)  # Dictionary to store numbers per coordinator
    pattern = r"Split [0-4]?[0-9]: (?:F1|Recall)\s+Coordinator\s+(\d+) => ([\d\.]+)"
    
    with open(filename, 'r') as file:
        for line in file:
            match = re.search(pattern, line)
            if match:
                coordinator = int(match.group(1))  # Extract Coordinator number
                value = float(match.group(2))      # Extract floating-point value
                data[coordinator].append(value)
    return data

            # Extract recalls for coordinators testing#
def extract_numbers_by_coordinator_testing(filename):
    data = defaultdict(list)  # Dictionary to store numbers per coordinator
    pattern = r"Split [5-9][0-9]: (?:F1|Recall)\s+Coordinator\s+(\d+) => ([\d\.]+)"
    
    with open(filename, 'r') as file:
        for line in file:
            match = re.search(pattern, line)
            if match:
                coordinator = int(match.group(1))  # Extract Coordinator number
                value = float(match.group(2))      # Extract floating-point value
                data[coordinator].append(value)
    return data
            # Extract recalls for coordinators voting again#
def extract_numbers_by_coordinator_voting2(filename):
    data = defaultdict(list)  # Dictionary to store numbers per coordinator
    pattern = r"Split 1[0-4][0-9]: (?:F1|Recall)\s+Coordinator\s+(\d+) => ([\d\.]+)"
    
    with open(filename, 'r') as file:
        for line in file:
            match = re.search(pattern, line)
            if match:
                coordinator = int(match.group(1))  # Extract Coordinator number
                value = float(match.group(2))      # Extract floating-point value
                data[coordinator].append(value)
    return data
    
# Extract recalls from models
def extract_numbers_by_coordinator_models(filename):
    data = defaultdict(list)  # Dictionary to store lists of values per coordinator
    pattern = r"Split \d+: (?:F1|Recall)\s+Models\s+Coordinator\s+(\d+) => \[([\d\.,\s]+)\]"

    with open(filename, 'r') as file:
        for line in file:
            match = re.search(pattern, line)
            if match:
                coordinator = int(match.group(1))  # Extract Coordinator number
                values = list(map(float, match.group(2).split(',')))  # Convert list elements to floats
                data[coordinator].append(values)
    return data

def calculate_column_means(data):
    means = {}

    for coordinator, lists in data.items():
        # Transpose list of lists to group values by column
        num_columns = len(lists[0])  # Assuming all rows have the same number of columns
        column_sums = [0] * num_columns
        column_counts = [0] * num_columns

        for row in lists:
            for i, value in enumerate(row):
                if value != 0:  # Ignore zeros
                    column_sums[i] += value
                    column_counts[i] += 1

        # Compute mean for each column (avoid division by zero)
        column_means = [
            (column_sums[i] / column_counts[i]) if column_counts[i] > 0 else 0
            for i in range(num_columns)
        ]

        means[coordinator] = column_means  # Store results

    return means
    
            # Extract recalls for coordinators models voting#
def extract_numbers_by_coordinator_models_voting(filename):
    data = defaultdict(list)  # Dictionary to store lists of values per coordinator
    pattern = r"Split [0-4]?[0-9]: (?:F1|Recall)\s+Models\s+Coordinator\s+(\d+) => \[([\d\.,\s]+)\]"

    with open(filename, 'r') as file:
        for line in file:
            match = re.search(pattern, line)
            if match:
                coordinator = int(match.group(1))  # Extract Coordinator number
                values = list(map(float, match.group(2).split(',')))  # Convert list elements to floats
                data[coordinator].append(values)
    return data

def extract_numbers_by_coordinator_models_testing(filename):
    data = defaultdict(list)  # Dictionary to store lists of values per coordinator
    pattern = r"Split [5-9][0-9]: (?:F1|Recall)\s+Models\s+Coordinator\s+(\d+) => \[([\d\.,\s]+)\]"

    with open(filename, 'r') as file:
        for line in file:
            match = re.search(pattern, line)
            if match:
                coordinator = int(match.group(1))  # Extract Coordinator number
                values = list(map(float, match.group(2).split(',')))  # Convert list elements to floats
                data[coordinator].append(values)
    return data

def extract_numbers_by_coordinator_models_voting2(filename):
    data = defaultdict(list)  # Dictionary to store lists of values per coordinator
    pattern = r"Split 1[0-4][0-9]: (?:F1|Recall)\s+Models\s+Coordinator\s+(\d+) => \[([\d\.,\s]+)\]"

    with open(filename, 'r') as file:
        for line in file:
            match = re.search(pattern, line)
            if match:
                coordinator = int(match.group(1))  # Extract Coordinator number
                values = list(map(float, match.group(2).split(',')))  # Convert list elements to floats
                data[coordinator].append(values)
    return data


def calculate_overall_column_mean(mean_values):
    num_columns = len(next(iter(mean_values.values())))  # Get number of columns from first coordinator
    overall_means = []

    for col in range(num_columns):
        column_values = [means[col] for means in mean_values.values() if means[col] != 0]  # Ignore 0s
        mean_value = sum(column_values) / len(column_values) if column_values else 0  # Avoid division by zero
        overall_means.append(mean_value)

    return overall_means
```

```python
# Recall results for A1, E1 without attack - All 

print("Results for coordinators Analysis 1, Experiment 1 - Without attack")
filename_A1E1 = "experiments/A1E1_NOATT.out"
data_coordinators = extract_numbers_by_coordinator(filename_A1E1)
coordinators_mean_values = calculate_means(data_coordinators)

data_models = extract_numbers_by_coordinator_models(filename_A1E1)
models_mean_values = calculate_column_means(data_models)
# Print results
print("Coordinators All")
for coordinator, means in sorted(coordinators_mean_values.items()):
    print(f"Coordinator {coordinator}: Column-wise Means = {means}")
print("Models All")
for coordinator, means in sorted(models_mean_values.items()):
    print(f"Coordinator {coordinator}: Column-wise Means = {means}")

print(f"Overall mean values for models: {calculate_overall_column_mean(models_mean_values)}")
print("")

# Recall results for A1, E1 without attack - Voting
print("Results for models Analysis 1, Experiment 1 - Without attack")
data_coordinators = extract_numbers_by_coordinator_voting(filename_A1E1)
coordinators_mean_values = calculate_means(data_coordinators)

data_models = extract_numbers_by_coordinator_models_voting(filename_A1E1)
models_mean_values = calculate_column_means(data_models)

# Print results
print("Coordinators Voting")
for coordinator, means in sorted(coordinators_mean_values.items()):
    print(f"Coordinator {coordinator}: Column-wise Means = {means}")
print("Models Voting")
for coordinator, means in sorted(models_mean_values.items()):
    print(f"Coordinator {coordinator}: Column-wise Means = {means}")

print(f"Overall mean values for models voting: {calculate_overall_column_mean(models_mean_values)}")
print("")

# Recall results for A1, E1 without attack - Testing
print("Results for models Analysis 1, Experiment 1 - Without attack")
data_coordinators = extract_numbers_by_coordinator_testing(filename_A1E1)
coordinators_mean_values = calculate_means(data_coordinators)

data_models = extract_numbers_by_coordinator_models_testing(filename_A1E1)
models_mean_values = calculate_column_means(data_models)

# Print results
print("Coordinators Testing")
for coordinator, means in sorted(coordinators_mean_values.items()):
    print(f"Coordinator {coordinator}: Column-wise Means = {means}")
print("Models Testing")
for coordinator, means in sorted(models_mean_values.items()):
    print(f"Coordinator {coordinator}: Column-wise Means = {means}")

print(f"Overall mean values for models testing: {calculate_overall_column_mean(models_mean_values)}")
print("")

# Recall results for A1, E1 without attack - Voting2
print("Results for models Analysis 1, Experiment 1 - Without attack")
data_coordinators = extract_numbers_by_coordinator_voting2(filename_A1E1)
coordinators_mean_values = calculate_means(data_coordinators)

data_models = extract_numbers_by_coordinator_models_voting2(filename_A1E1)
models_mean_values = calculate_column_means(data_models)

# Print results
print("Coordinators Voting2")
for coordinator, means in sorted(coordinators_mean_values.items()):
    print(f"Coordinator {coordinator}: Column-wise Means = {means}")
print("Models Voting2")
for coordinator, means in sorted(models_mean_values.items()):
    print(f"Coordinator {coordinator}: Column-wise Means = {means}")

print(f"Overall mean values for models voting2: {calculate_overall_column_mean(models_mean_values)}")
print("")
```
